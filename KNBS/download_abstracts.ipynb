```python
import os
import requests
import urllib3

# Disable SSL warnings
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# === Step 1: Set up paths ===
txt_file = r"C:\Users\Rono\Downloads\AAAA Hackathon\Datasets\statistical_abstracts.txt"
output_dir = r"C:\Users\Rono\Downloads\AAAA Hackathon\Datasets\County_statistical_abstracts"
os.makedirs(output_dir, exist_ok=True)

# === Step 2: Read URLs ===
with open(txt_file, "r", encoding="utf-8") as f:
    urls = [line.strip() for line in f if line.strip().lower().endswith(".pdf")]

print(f"üîç Found {len(urls)} PDF URLs in file.")

# === Step 3: Download each PDF ===
for url in urls:
    try:
        filename = os.path.basename(url)
        output_path = os.path.join(output_dir, filename)

        # Skip if already downloaded
        if os.path.exists(output_path):
            print(f"‚úÖ Already exists: {filename}")
            continue

        print(f"‚¨áÔ∏è Downloading: {filename}")
        response = requests.get(url, stream=True, timeout=30, verify=False)

        if response.status_code == 200:
            with open(output_path, 'wb') as f:
                for chunk in response.iter_content(chunk_size=8192):
                    f.write(chunk)
            print(f"‚úÖ Saved: {filename}")
        else:
            print(f"‚ùå Failed: {filename} - HTTP {response.status_code}")

    except Exception as e:
        print(f"‚ùå Error downloading {url}: {e}")
```
